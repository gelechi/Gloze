# Data access - Beautiful Soup

```python
url = "https://en.wikipedia.org/wiki/Beautiful_Soup_(HTML_parser)"

response = requests.get(url)
```

```python
response.text
#when we read a website (html file) in as text, it is uninterpretable and hard to read.
#it gets displayed as one long string, which doesnt help us retrieve the info we need
```

```python
from bs4 import BeautifulSoup

#this library will help us make sense of the incoming response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup.find("h1")
```

```python
soup.find("h1").text
```

```python
h2s = soup.find_all("h2")
h2s
```

```python
for tag in h2s:
  print(tag.text)
```

```python
soup.find("li", {"id":"ca-talk"}).text
```

```python
soup.find("li", {"id":"ca-talk"}).find("a").get("href")
```
